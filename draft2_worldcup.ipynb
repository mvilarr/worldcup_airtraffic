```python
import requests
```

```python
#scraping fifa website to get where the world cup matches are going to happen
url = "https://www.fifa.com/en/tournaments/mens/worldcup/canadamexicousa2026/articles/match-schedule-fixtures-results-teams-stadiums"
resp = requests.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
#<span title="&quot;Thursday, 11 June 2026&quot;" style="display: -webkit-box; -webkit-line-clamp: 5; -webkit-box-orient: vertical; overflow: hidden;"><strong>Thursday, 11 June 2026</strong></span>
```

```python
# read the webpage as bs4
soup = BeautifulSoup(resp.text)

# select all the "a" tags with the specified class.
articles = soup.find_all("h4", {"class": "span-title"})

# iterate through each headline and grab the title and link of each story.
data = []
for elm in articles:
    link = elm.get('href')
    title = elm.text
    row = {'link' : link, 'title': title}
    data.append(row)
    
data[:5]
```
