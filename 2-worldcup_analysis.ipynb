## geocoding airports

```python
load_dotenv()
```

```python
api_key_geocoder = os.environ["api_key_geocoder"]
```

```python
api_key_geocoder[:9]+"..."
```

```python
top_50_airports = pd.read_csv('top50_airports.csv')
```

```python
top_50_airports.head(5)
```

```python
#getting the latitude and longitude
for index, row in top_50_airports.iterrows():
    full_address = f"{row['ICAO']}, {row['Airport']}, {row['City/Country']}"
    print(full_address)
    
    geo = geocoder.google(full_address, key=api_key_geocoder)
    print("Status:", geo.status)
    print("OK:", geo.ok)
    print("LatLng:", geo.latlng)
    
    if geo.ok:
        top_50_airports.loc[index, "lat"] = geo.latlng[0]
        top_50_airports.loc[index, "long"] = geo.latlng[1]
    else:
        print(f"Failed for {full_address}: {geo.status}")

top_50_airports
```

```python
top_50_airports = top_50_airports.rename(columns={"Airport":"dep_airport",
                                "lat":"lat_dep",
                                "long": "long_dep"})
top_50_airports
```

```python
top_50_airports = top_50_airports.drop(columns=["City/Country"])
top_50_airports
```

```python
top_50_airports["arrival"] = "New York"
```

```python
top_50_airports.head(5)
```

```python
top_50_airports ["lat_ny"] = 40.7128
```

```python
top_50_airports ["long_ny"] = -74.0060
```
